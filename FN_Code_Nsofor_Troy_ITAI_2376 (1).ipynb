```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "microsoft/Phi-3-mini-4k-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

def local_llm(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs, max_new_tokens=300)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
from langchain.llms.base import LLM

class LocalLLMWrapper(LLM):
    model_name = "custom-phi-3"

    def __init__(self, model_fn, **kwargs):
        super().__init__(model_name=self.model_name, **kwargs)
        self.model_fn = model_fn

    def _call(self, prompt, stop=None):
        return self.model_fn(prompt)

    @property
    def _identifying_params(self):
        return {"model_fn_name": self.model_fn.__name__ if hasattr(self.model_fn, '__name__') else "anonymous_fn",
                "model_name": self.model_name}

    @property
    def _llm_type(self):
        return "local_hf_wrapper"

    # Keeping original run and predict methods for potential other uses.
    def run(self, prompt):
        return self.model_fn(prompt)

    def predict(self, prompt):
        return self.model_fn(prompt)
```

```python
from langchain.llms.base import LLM

class DummyLLM(LLM):
    def _call(self, prompt, stop=None):
        return "OK"

    @property
    def _identifying_params(self):
        return {}

    @property
    def _llm_type(self):
        return "dummy"
```

```python
import os
import logging
import sys
from crewai_tools import tool
from crewai import Agent, Task, Crew, Process

# Use your local LLM wrapper
# llm = None # Removed this line as it was causing issues

# TOOL: Wrap your local HuggingFace model
@tool
def local_model_tool(prompt: str) -> str:
    """Run the local HuggingFace model on the given prompt."""
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs, max_new_tokens=300)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# -------------------------
# AGENTS
# -------------------------

weather_agent = Agent(
    name="Weather Assistant",
    role="Provides accurate and concise weather updates",
    goal="Provide current weather and a short forecast for a given location",
    backstory="A seasoned meteorologist AI.",
    instructions="Provide accurate and concise weather updates.",
    verbose=False,
    allow_delegation=False,
    llm=DummyLLM(),
    tools=[local_model_tool]
)

extraction_agent = Agent(
    role="Extraction Agent",
    goal="Extract requirements from messy instructions",
    backstory="A skilled AI assistant specialized in dissecting complex documents.",
    llm=DummyLLM(),
    tools=[local_model_tool]
)

planning_agent = Agent(
    role="Planning Agent",
    goal="Break requirements into tasks",
    backstory="An expert in project management.",
    llm=DummyLLM(),
    tools=[local_model_tool]
)

# -------------------------
# TASKS
# -------------------------

extract_task = Task(
    description="Extract requirements from the input text.",
    agent=extraction_agent,
    expected_output="A clear and concise list of all distinct requirements.",
    llm=DummyLLM()
)

plan_task = Task(
    description="Create a task plan based on extracted requirements.",
    agent=planning_agent,
    expected_output="A structured task plan.",
    llm=DummyLLM()
)

weather_task = Task(
    description="Find the current weather and a short forecast.",
    agent=weather_agent,
    expected_output="Weather conditions and a brief forecast.",
    llm=DummyLLM()
)

# -------------------------
# CREWS
# -------------------------

weather_crew = Crew(
    agents=[weather_agent],
    tasks=[weather_task],
    verbose=False,
    llm=DummyLLM()
)


crew = Crew(
    agents=[extraction_agent, planning_agent],
    tasks=[extract_task, plan_task],
    llm=DummyLLM()
)

# -------------------------
# RUNNER
# -------------------------

def run_weather_crew(user_input: str, output_file=None, suppress_output=False):
    if suppress_output:
        original_stdout = sys.stdout
        original_stderr = sys.stderr
        sys.stdout = open(os.devnull, 'w')
        sys.stderr = open(os.devnull, 'w')

    try:
        result = weather_crew.kickoff(inputs={'location': user_input})
        if output_file:
            with open(output_file, 'w') as f:
                f.write(result)
        return result
    finally:
        if suppress_output:
            sys.stdout.close()
            sys.stderr.close()
            sys.stdout = original_stdout
            sys.stderr = original_stderr

# Example
run_weather_crew("What is the weather in Seattle?")
```
